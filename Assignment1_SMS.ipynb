```python
import matplotlib
from pylab import plot, show, plt


a = 6
b = -4 

def initialize():
    global x, result
    x = 1.
    result = [x]
    
def observe():
    global x, result
    result.append(x)
    
def update():
    global x, result
    x = (a * x) + b
    
initialize()
for t in range(30):
    update()
    observe()
    
print(result)

#plot the results
plot(result, color = 'plum', marker = '*', label = 'a = 6, -4')
plt.legend()
show()
```
