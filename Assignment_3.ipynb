```python
from pylab import *

a = b = c = d = 1
deltat1 = 0.01
deltat2 = 0.003

def initialize():
    global x1, y1, x2, y2, t, result1, result2, timesteps
    x1 = x2 = 0.1
    y1 = y2 = 0.1
    t = 0.0
    result1 = [x1]
    result2 = [x2]
    timesteps = [t]
    
def observe():
    global x1, y1, x2, y2, t, result1, result2, timesteps
    result1.append(x1)
    result2.append(x2)
    timesteps.append(t)
  
def update():
    global x1, y1, x2, y2
    nextx1 = x1 + (a * x1 - b * x1 * y1) * deltat1
    nexty1 = y1 + (- c * y1 + d * x1 * y1) * deltat1
    x1, y1 = nextx1, nexty1
    
    nextx2 = x2 + (a * x2 - b * x2 * y2) * deltat2
    nexty2 = y2 + (- c * y2 + d * x2 * y2) * deltat2
    x2, y2 = nextx2, nexty2
    
initialize()

while t < 50:
    update ()  
    t += deltat1
    observe()
    
plt.plot(timesteps, result1, label = 'deltat = {0.01}')

initialize()

timesteps.clear()
result2.clear()
while t < 50:
    update ()  
    t += deltat2
    observe()

plt.plot(timesteps, result2, label = 'deltat = {0.003}')
xlabel('Time')
ylabel('Population')
legend()  
show()    
```
